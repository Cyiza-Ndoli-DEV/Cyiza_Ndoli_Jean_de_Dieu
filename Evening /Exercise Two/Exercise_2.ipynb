```python
# 5
import pandas as pd

# Load the DataFrame from the CSV file
df = pd.read_csv('/home/cheguezulu/Jupyter_Work/Exercise Two/AfricaCupofNationsMatches.csv')

# Find the most frequent value in the 'HomeTeamGoals' column
most_frequent = df['HomeTeamGoals'].mode()[0]

# Replace all other values with 'Other'
df['HomeTeamGoals'] = df['HomeTeamGoals'].apply(lambda x: x if x == most_frequent else 'Other')

# Display the modified 'HomeTeamGoals' column
print(df['HomeTeamGoals'])


```
